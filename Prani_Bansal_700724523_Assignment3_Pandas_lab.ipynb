### Task 13
Using three courses' scores (assuming each course has 3 credits), calculate each student's GPA for Spring 2022, respectively. The following criteria should be used in the calculation
+ `A: 4.0`
+ `B: 3.0`
+ `C: 2.0`
+ `D: 1.0`
+ `F: 0.0`

For example, if a student's grades are A, B and C. That student's GPA should be calculated as

`(3 * 4 + 3 * 3 + 3 * 2) / 9 = 3.0`

Add a new column titled `GPA` in the scores DataFrame to store the GPA for each student

```python
david_combine_scores_df['GPA']=''
def calc_gpa(score):
    if (score >= 90): return 4
    if (80 <= score < 90): return 3
    if  (70 <= score < 80): return 2
    if (60 <= score < 70): return 1
    if (score < 60):       return 0

   
for i in range(11):
    a=david_combine_scores_df.iloc[i][0]
    b=david_combine_scores_df.iloc[i][1]
    c=david_combine_scores_df.iloc[i][2]
   
    sum=(calc_gpa(a)+calc_gpa(b)+calc_gpa(c))/3
    david_combine_scores_df.iloc[i, david_combine_scores_df.columns.get_loc('GPA')] = sum
david_combine_scores_df
```

### Task 14
Insert a new column title `Status` to provide a student academic standing based on the GPA
+ `Honor Roll: GPA >= 3.5`
+ `Normal: 3.0 <= GPA < 3.5`
+ `In Probation: GPA < 3.0`

```python
david_combine_scores_df['Status'] = ''
david_combine_scores_df.loc[david_combine_scores_df['GPA']>= 3.5, 'Status'] = 'Honor Roll'
david_combine_scores_df.loc[(david_combine_scores_df['GPA']>=3) & (david_combine_scores_df['GPA'] < 3.5), 'Status'] = 'Normal'
david_combine_scores_df.loc[(david_combine_scores_df['GPA'] < 3), 'Status'] = 'In Probation'

david_combine_scores_df
```
