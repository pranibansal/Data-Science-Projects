### Pandas Built-in Visualization Tools
`DataFrame.plot` can be used to make plots of `Series` or `DataFrame`. By default, matplotlib is used

<div>

</div>

```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips['total_bill'].plot(kind = 'box' )
```

```python
tips['total_bill'].plot.box()
```

```python
tips.plot.box()
```

```python
tips['tip'].plot(kind = 'hist')
```

```python
tips['tip'].plot.hist()
```

```python
tips['tip'].plot(kind = 'hist', bins = 20)
```

```python
tips.groupby('sex')['total_bill'].mean().plot.bar()
```

```python
tips.pivot_table(index = 'day', columns = 'sex', values = 'total_bill').plot.bar()
```

```python
df = pd.DataFrame(np.random.randn(1000, 3), columns = ['A', 'B', 'C'])
df.head()
```

```python
df.plot.scatter(x = 'A', y = 'B')
```

```python
df.plot.scatter(x = 'A', y = 'B', c = 'C', figsize = (10, 6))
```

```python
df.plot.scatter(x = 'A', y = 'B', s = df['C'] * 60, figsize = (10, 6))
```

```python
df.plot.box()
```

```python
np.abs(df.iloc[:6]).plot.bar(figsize = (10, 6))
```

```python
np.abs(df.iloc[:6]).plot.bar(figsize = (10, 6), stacked = True)
```

```python
df.plot.hexbin(x = 'A', y = 'B', figsize = (10, 6), gridsize = 30, cmap = 'coolwarm')
```

```python
df['A'].plot.kde()
```

```python
df['A'].plot.density()
```

```python
df.plot.density()
```
